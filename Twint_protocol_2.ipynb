```python
#Imports
import twint
import nest_asyncio
nest_asyncio.apply()
import pandas as pd
```

```python
# Configure
c = twint.Config()

#Set search conditions
c.Search = "Ethereum"
c.Since = '2018-1-1'
c.Until = '2018-08-11'
c.Pandas = True               #This enables pandas for the function below
c.Resume = 'eth_last.csv'     #This file saves the last get call so that if you have to restart it won't restart from the beginning
c.Output = "2018.csv"         #This creates a csv file. I don't like the file but it saves it as backup.

# Run the search
twint.run.Search(c)           #This runs twint
```

```python
#A function that returns the column names of the Pandas dataframe
def column_names():
    return twint.output.panda.Tweets_df.columns

#A function that creats a dataframe having only the columns that we are interested in.
def twint_to_pd(columns):
    return twint.output.panda.Tweets_df[columns]
```

```python
column_names()      #To get an idea of what is downloaded
```

```python
#Creates a data frame, select the columns that you want
df = twint_to_pd(["id", "date", "tweet", "language", "hashtags", "username", "nlikes", "nretweets"])
```
