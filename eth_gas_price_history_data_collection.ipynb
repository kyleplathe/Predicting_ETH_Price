```python
# Append url for our api
url = "https://owlracle.info/eth/history?candles=1000&timeframe=1d"
```

```python
# Read json data from API URL
eth_gas_history = pd.read_json(url, convert_dates = True)
```

```python
# View historical data
eth_gas_history
```

```python
# Separate gas price columns from json
gas_price_candles = pd.json_normalize(eth_gas_history['gasPrice'])
gas_price_candles
```

```python
# Concat gas_price_candles and eth_gas_history to a new dataframe
gas_prices_df = pd.concat([gas_price_candles, eth_gas_history], axis=1)

# Drop gasPrice Column
gas_prices_df.drop(['gasPrice','samples'], axis=1, inplace=True)

# Set timestamp as index
gas_prices_df = gas_prices_df.set_index("timestamp")

# Add formated date column
gas_prices_df['date'] = gas_prices_df.index.strftime("%Y-%m-%d")

# Set index to date column
gas_prices_df = gas_prices_df.set_index('date')
gas_prices_df
```

```python
# Writing my gas_prices_df to CSV file
gas_prices_df.to_csv("./eth_gas_prices_1.csv")
```
