```python
#Imports
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#Read in the csv file
df = pd.read_csv('/Volumes/5Tb/Test/Project_2/Resources/2019_file_master_en.csv', index_col = "date", infer_datetime_format=True, parse_dates=True)

#df = df.drop(columns = ["Column 1", "id", "language"])

#Renaming the dataframe to identify it as a pandas dataframe (when using with Optimus)
df_pd = df
df_pd
```

```python
df_pd.info()
```

```python
# Download/Update the VADER Lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#List comprehension to create the table.
df_pd['compound'] = [analyzer.polarity_scores(x)['compound'] for x in df_pd['tweet']]
df_pd['neg'] = [analyzer.polarity_scores(x)['neg'] for x in df_pd['tweet']]
df_pd['neu'] = [analyzer.polarity_scores(x)['neu'] for x in df_pd['tweet']]
df_pd['pos'] = [analyzer.polarity_scores(x)['pos'] for x in df_pd['tweet']]

#Look at the dataframe
df_pd
```
